```python
import numpy as np
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/SpotifyFeatures.csv')
```

```python
data
```

```python
from sklearn.neighbors import KNeighborsClassifier

x = data["danceability"]
y = data["energy"]
classes = data["popularity"]

data = list(zip(x, y))
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(data,classes)
```

```python
import matplotlib.pyplot as plt

nx = 0.607
ny = 0.706

new_point = [(nx,ny)]

prediction = knn.predict(new_point)

plt.scatter(x + [nx], y + [ny], c=classes + [prediction[0]])
plt.text(x=nx-1.7, y=ny-0.7, s=f"new point, class: {prediction[0]}")

plt.scatter([nx],[ny],[prediction[0]])
plt.show()
```
